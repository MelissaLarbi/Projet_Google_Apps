```python
def func(i):
    """
    une fonction qui va nous aider dans le processus du data cleaning 
    """
    try:
        i = i.replace("+", '') 
        i = i.replace(",", '')
        
        if 'm' in i:
            i = i.replace('m', '')
            d = float(i)*10**6
        
        elif 'k' in i:
            i = i.replace('k', '')
            d = float(i)*10**3
        else:
            d = float(i)*10**3
        
        return d
    except Exception as e :
        return i
```

# **<font color="RED"> 12 ) L'impact de la catégorie sur le taux d'istallations </font>** 

```python
nb_install  = []
categories= ["BUSINESS","FAMILY","GAME","PERSONALIZATION","LIFESTYLE","FINANCE","EDUCATION","MEDICAL","TOOLS","PRODUCTIVITY"]
groups = googleplaystore.groupby(["Category"]) # on regroupe le dataset selon les catégories 

for categorie in categories: # pour chaque catégorie 
    nb_install.append((categorie,groups.get_group(categorie)["Installs"].mean())) # on ajoute le nom de la catégories et la moyenne de ses installations

bests= sorted(nb_install, key=lambda x: x[1])[0:5] # trier selon da moyenne des installations et on prend les 5 première 
bests_cat = [c[0] for c in bests]  # on récupère juste le nom des best catégories 
bests_notes = [c[1] for c in bests] # on récupère la moyenne des installation des best catégories

fig = plt.figure()
width = 0.5
plt.bar(bests_cat, bests_notes)
```

#### Commentaire: 

   On voit que la catégorie la plus installer est FAMILLY avec une moyenne de plus de 2.3x10e6 des instalations
   puis la catégorie BUSINESS avec une moyenne de plus de 1.5x10e6 installations

# **<font color="RED"> 15) Gratuit VS payant: Taux d'installation</font>**
